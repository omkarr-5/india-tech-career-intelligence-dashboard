```python
import pandas as pd
df = pd.read_csv("india_job_market_2024_2026.csv")

skills_df = (
    df.assign(
        Skills_Required=df['Skills_Required'].str.split(',')
    )
    .explode('Skills_Required')
)

skills_df['Skills_Required'] = skills_df['Skills_Required'].str.strip()

skills_df.head()
```
